```python
from random import shuffle

q = 5
n = 6
k = 2
print("q = ", q)
print("n = ", n)
print("k = ", k)
E = VectorSpace(GF(q), n)

dgv = 0
while(binomial(n,dgv)*((q-1)^dgv)/(q^(n-k)) < 1):
    dgv += 1
print("dgv = ", dgv)
w = dgv
w = 2
print("w = ", w)

nb_tests = 1000
count_C = 0
count_coset = 0

for i in range(nb_tests):
    L = [E.random_element() for _ in range(k)]
    C = span(L, GF(q))
    
    e_tab = [0]*(n-w)
    for _ in range(w):
        x = GF(q).random_element()
        while x == 0:
            x = GF(q).random_element()
        e_tab += [x]
    shuffle(e_tab)
    e = vector(GF(q), e_tab)

    for c in C:
        if c.hamming_weight() == w:
            count_C += 1
        if (e + c).hamming_weight() == w:
            count_coset += 1
    
print("count_C = ", float(count_C/nb_tests), " = ", float(binomial(n, w) * (q-1)^(w) / q^(n-k)))
print("count_coset = ", float(count_coset/nb_tests), " = ", float(1 + (binomial(n, w) * (q-1)^(w) - 1) / q^(n-k)))
```
